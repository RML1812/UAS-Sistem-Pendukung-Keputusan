```python
features = {
    'trackFeatureAcousticness': 'acousticness',
    'trackFeatureDanceability': 'danceability',
    'trackFeatureEnergy': 'energy',
    'trackFeatureInstrumentalness': 'instrumentalness',
    'trackFeatureLiveness': 'liveness',
    'trackFeatureLoudness': 'loudness',
    'trackFeatureSpeechiness': 'speechiness',
    'trackFeatureTempo': 'tempo',
    'trackFeatureValence': 'valence'
}
```

```python
def euclideanDistance(data):
    return sqrt(sum([i ** 2 for i in data]))
```

```python
# from every feature, print the max and min value
for key, value in features.items():
    print('Max {} is {}'.format(value, korea[key].max()))
    print('Min {} is {}'.format(value, korea[key].min()))
```

```python
# Topsis Algorithm for music recommendation based on the 'trackFeatureAcousticness', 'trackFeatureDanceability', 'trackFeatureEnergy', 'trackFeatureInstrumentalness', 'trackFeatureLiveness', 'trackFeatureLoudness', 'trackFeatureSpeechiness', 'trackFeatureTempo'
weight = [1, 1, 1, 1, 1, 1, 1, 1, 1]
# impact = [0, 0, 0, 0, 0, 0, 0, 0]
impact = [1, 1, 1, 1, 1, 1, 1, 1, 1]

# Step 1: Normalize the data
for key, value in features.items():
    korea[value + 'Normalized'] = korea[key] / euclideanDistance(korea[key])
    # Step 2: Calculate the weighted normalized decision matrix
    korea[value + 'Normalized'] *= weight[list(features.keys()).index(key)] 

# Step 3: Determine the ideal and negative-ideal solution
ideal = {}
negativeIdeal = {}
for key, value in features.items():
    if impact[list(features.keys()).index(key)] == 1:
        ideal[value] = korea[value + 'Normalized'].max()
        negativeIdeal[value] = korea[value + 'Normalized'].min()
    elif impact[list(features.keys()).index(key)] == 0:
        ideal[value] = korea[value + 'Normalized'].min()
        negativeIdeal[value] = korea[value + 'Normalized'].max()

# Step 4: Calculate the separation measures from the sum of all features
for key, value in features.items():
    korea[value + 'SeparationMeasureIdeal'] = pow(korea[value + 'Normalized'] - ideal[value], 2)
    korea[value + 'SeparationMeasureNegativeIdeal'] = pow(korea[value + 'Normalized'] - negativeIdeal[value], 2)

# Sum separation measures for all features
korea['SeparationMeasureIdeal'] = korea[[f'{value}SeparationMeasureIdeal' for value in features.values()]].sum(axis=1)
korea['SeparationMeasureNegativeIdeal'] = korea[[f'{value}SeparationMeasureNegativeIdeal' for value in features.values()]].sum(axis=1)

# Take the square root of the sum
korea['SeparationMeasureIdeal'] = korea['SeparationMeasureIdeal'].apply(sqrt)
korea['SeparationMeasureNegativeIdeal'] = korea['SeparationMeasureNegativeIdeal'].apply(sqrt)

# Step 5: Calculate the relative closeness to the ideal solution
korea['totalRelativeCloseness'] = korea['SeparationMeasureNegativeIdeal'] / (korea['SeparationMeasureNegativeIdeal'] + korea['SeparationMeasureIdeal'])

# Step 6: Rank the relative closeness and print top 10
korea = korea.sort_values(by=['totalRelativeCloseness'], ascending=False)

# Step 7: Print 5 recommended songs, sorted by totalRelativeCloseness
korea[['trackName', 'artistName', 'trackUrl', 'totalRelativeCloseness']].head(15)
```
